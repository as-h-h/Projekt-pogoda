```python
import os
import time
import pandas as pd
```

Pobieranie plików ze strony

```python
os.mkdir("C:/Users/Dell/Desktop/Pobrane_do_projektu")

for year in range(2001, 2024):
    for month in range(1, 13):
        output_dir = f"C:\\Users\\Dell\\Desktop\\Pobrane_do_projektu"

        url = f"https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteorologiczne/dobowe/klimat/{year}/{year}_{month:02d}_k.zip"

        print("Pobiera url",url)

        os.system(f'powershell Invoke-WebRequest -Uri {url} -OutFile "{year}_{month}.zip"')
        os.system(f'powershell Expand-Archive -Path "{year}_{month}.zip" -Force -DestinationPath {output_dir}')
        for file in os.listdir(f"C:\\Users\\Dell\\Desktop\\Pobrane_do_projektu"):
            if file.startswith('k_d_t'):
                os.remove(f"C:\\Users\\Dell\\Desktop\\Pobrane_do_projektu\\k_d_t_{month:02d}_{year}.csv")
        #for file in os.listdir(f"C:\\Users\\Dell\\PycharmProjects\\PythonProject\\Warsztat AI\\Projekt"):
        #    os.remove(f"C:\\Users\\Dell\\PycharmProjects\\PythonProject\\Warsztat AI\\Projekt\\{year}_{month}.zip")


        time.sleep(1)
```

Scalanie plików w jeden

```python
base_dir = r"C:\Users\Dell\Desktop\Projekt_PWAI\Pobrane_do_projektu"
output = r"C:\Users\Dell\Desktop\Projekt_PWAI\Dane_meteorologiczne.csv"

first = True

for year in range(2001, 2024):
    for month in range(1, 13):
        file = f"k_d_{month:02d}_{year}.csv"
        dire = os.path.join(base_dir, file)

        if os.path.exists(dire):
            df = pd.read_csv(dire, encoding='cp1250')

            df.to_csv(
                output,
                mode="w" if first else "a",
                index=False,
                header=False,
                encoding='utf-8'
            )
            first = False
```

Usuwanie zbędnych kolumn

```python
df = pd.read_csv(output, header=None, usecols=[1, 2, 3, 4, 5, 7, 9])
df.to_csv(
                output,
                mode="w",
                index=False,
                header=False,
            )
```
