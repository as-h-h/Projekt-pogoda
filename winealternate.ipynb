```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/GJ-Demo/PWAI/refs/heads/main/Wyk%C5%82ad12/wine.txt")

cols = df.columns
cols = cols[cols != "class"]

for col1 in cols:
    for col2 in cols:
        if col1 == col2:
            continue
        else:
            #sns.set_palette("Spectral")
            #sns.set_style("darkgrid")
            sns.set_theme(context="talk", style="dark", palette="pastel")
            sns.scatterplot(
            data = df,
            x=col1,
            y=col2,
            hue="class"
            )
            plt.show()
```
